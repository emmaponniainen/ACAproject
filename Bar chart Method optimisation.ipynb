```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sc
from scipy.ndimage.filters import gaussian_filter1d
from scipy.interpolate import make_interp_spline, BSpline

fig1 = plt.figure(figsize=(4,4))
ax1=fig1.add_subplot(1,1,1)


#read datafile with half lives
df =  pd.read_csv(r'Half lives')

print(df)

#Extracting the first or second data into two separate numpy arrays
data = np.array(df)
first = data[::2][:,2]

mean= np.mean(first)
print(first)
print(first.std())

std=first.std()

second = data[1::2][:,2]
print(mean)


width=10

#Make the bar charts
ax1.bar(0,mean, width, color="grey", edgecolor="black",fill=False, hatch='//', yerr=std, capsize=5, label="$1^{st}$ closing")
ax1.bar([30,45,60,120], second, width, color="grey", edgecolor="black", label = "$2^{nd}$ closing")

ax1.set_xlim(-10,130)


ax1.set_ylabel('$t_{1/2}$ (min)', size=15)
ax1.set_xlabel('Light interval (min)', size=15)

plt.legend(frameon=False)

ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)


plt.xticks([0,30,45,60,120], ['0','30','45','60','120'], size=12)


#saving the figure
#plt.savefig('Graphs/.png', dpi=300, bbox_inches="tight")
```
